```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
image=cv2.imread('2.jpg')

gray=cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
#cv2.imshow('original image',image)
ret, thresh=cv2.threshold(gray,176,255,0)
contours, hierarchy = cv2.findContours(thresh.copy(), cv2.RETR_CCOMP, cv2.CHAIN_APPROX_NONE)
n=len(contours)-1
contours=sorted(contours,key=cv2.contourArea,reverse=False)[:n]

#print(contours)
print('Numbers of contours found=' + str(len(contours)))
```

```python
for c in contours:
    hull=cv2.convexHull(c)
    cv2.drawContours(image,[hull],0,(0,255,0),2)
cv2.imshow('convex hull',image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import cv2

image = cv2.imread('2.jpg')
edged = cv2.Canny(image, 175, 200)

contours, hierarchy = cv2.findContours(edged, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
cv2.drawContours(image, contours, -1, (0,255,0), 1)

cv2.imshow("Show contour", image)
cv2.waitKey(0)
cv2.destroyAllWindows()

for i,c in enumerate(contours):
    rect = cv2.boundingRect(c)
    x,y,w,h = rect
    box = cv2.rectangle(image, (x,y), (x+w,y+h), (0,0,255), 1)
    cropped = image[y: y+h, x: x+w]    
    #cv2.imshow("Show Boxes", cropped)
    #cv2.waitKey(0)
    #cv2.destroyAllWindows()
    cv2.imwrite("hull"+str(i)+".jpg", cropped)


cv2.imshow("Show Boxes", image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
